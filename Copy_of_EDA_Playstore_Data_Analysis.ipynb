## 3. ***Data Wrangling***

The datatype for columns like Reviews, Size, Installs, and price should be an int or float, but we can see that they are of object type here. Let's change them to the appropriate type.

### Reviews

```python
#converting the column Reviews to type 'int'
df1['Reviews'] = df1.Reviews.astype(int)
```

```python
df1.Reviews
```

### Size


It also includes the values "Varies with device", 'M' and 'k' which are present in the column 'size' as string in the value type. The KB and MB size scales would be messed up if the 'M' and 'K' were left out. Consequently, we convert KBs to MBs. Below is a method of doing this.Answer Here.

```python
#Removing the +Symbol:
df1['Size'] = df1.Size.apply(lambda x: x.strip('+'))
```

```python
#Removing the , symbol:
df1['Size'] = df1.Size.apply(lambda x: x.replace(',', ''))
```

```python
#removing M and k from values, also coverting KB into MB
df1['Size'] =df1['Size'].apply(lambda x: x.replace('M', '') if 'M' in str(x) else x )
df1['Size'] = df1['Size'].apply(lambda x: float(x.replace('k', ''))/1024 if 'k' in str(x) else x)
```

```python
#Replacing the Varies with device value with Nan :
df1['Size'] = df1.Size.replace('Varies with device', np.NaN)
```

We need to do something with the set of Nan values data since we converted the Varies with device value to Nan. Since some apps' sizes will be too huge and others excessively small, it would be best to remove the rows of the column Size that contain Nanvalues rather than attempt to replace them with mean or mode.

```python
# Removing the rows which containing "Varies with device"
df1.dropna(subset = ['Size'], inplace=True)
```

```python
#Renaming the column with appropiate name.
df1.rename(columns={'Size': 'Size(in MB)'},
          inplace=True)
```

```python
#Now, finally converting all these values to numeric type:
df1['Size(in MB)'] = pd.to_numeric(df1['Size(in MB)'])
```
